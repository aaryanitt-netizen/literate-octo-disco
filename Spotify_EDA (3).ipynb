CHanging the index column to relese_date column for our convenience 

```python
df_tracks.columns = df_tracks.columns.str.strip()
df_tracks.set_index("release_date",inplace= True)
#df_tracks.index=pd.to_datetime(df_tracks.index)  #not required cause its already in date time format in the file
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[5]
```

Changing the song duration from milli seconds to seconds

```python
df_tracks["Duration"]=df_tracks["duration_ms"].apply (lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace= True, axis=1)
df_tracks.head()
```

Correlation heat-map 

```python
corr_df = df_tracks.select_dtypes(include='number').corr(method='pearson')
#corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno',linewidths=1,linecolor='Black')
heatmap.set_title("Correlation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

sampling 0.4% of the data

```python
sample_df= df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

Regression line 

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",scatter_kws={"color": "#1DB954", "alpha": 0.5, "s": 30},
    line_kws={"color": "black", "linewidth": 1}).set(title="Loudness vs energy correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="acousticness",x="energy",scatter_kws={"color": "#1DB954", "alpha": 0.5, "s": 30},
    line_kws={"color": "black", "linewidth": 2}).set(title="acousticness vs energy correlation")
```

Line plot for number of songs released per year

```python
df_tracks['dates']=df_tracks.index
df_tracks['dates'] = pd.to_datetime(df_tracks.index,format='mixed',errors='coerce')
df_tracks['year'] = df_tracks.dates.dt.year
df_tracks = df_tracks[
    (df_tracks['year'] >= 1950) &
    (df_tracks['year'] <= 2023)]
songs_per_year = (df_tracks.groupby('year').size().reset_index(name='song_count'))
sns.lineplot(data=songs_per_year,x='year',y='song_count')
plt.title("Number of Songs Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Songs")
plt.show()
```

Duration of songs over the years

```python
year = df_tracks.dates.dt.year
total_dr=df_tracks.Duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
sns.barplot(
    x=year,
    y=total_dr,
    hue=year,            
    errorbar=None,
    legend=False,         
    ax=ax
)

ax.set_title("Year vs Duration")
plt.xticks(rotation=90)
plt.show()
```

Duration vs year line plot

```python
total_dr=df_tracks.Duration
sns.set_style(style='whitegrid')
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
sns.lineplot(
    x=year,
    y=total_dr,         
    ax=ax
)

ax.set_title("Year vs Duration")
plt.xticks(rotation=60)
plt.show()
```

Loading the Spotify genres dataset into a pandas DataFrame for analysis.

```python
df_genre=pd.read_csv(r"C:\Users\HP\Downloads\archive\SpotifyFeatures.csv")
```

```python
df_genre.head()
```

Finding song duration based on genres

```python
df_genre["Duration"]=df_genre["duration_ms"].apply (lambda x: round(x/1000))
df_genre.drop("duration_ms", inplace= True, axis=1)
plt.title("genre vise song durations")
palette = sns.color_palette("crest", n_colors=df_genre['genre'].nunique())
sns.barplot(y='genre',x='Duration', data=df_genre, hue='genre',palette=palette)
plt.xlabel("Duration in seconds")
plt.ylabel("Genres")
```

Popularity trend based on genres

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
palette = sns.color_palette("pastel", n_colors=df_genre['genre'].nunique())
famous=df_genre.sort_values("popularity", ascending=False).head(20)
sns.barplot(y='genre',x='popularity',data= famous,hue= 'genre', palette=palette).set(title="genre vise popularity trend")
```
